## DATA CLEANING

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(data={
    'Temperature':[25,np.nan,34,20,15,19,18,np.nan,25],
    'Humidity':[20,20,12,np.nan,13,15,np.nan,np.nan,20],
    'Situation':['Humid',np.nan, 'Too-Sunny','cool','Too Cool','Normal','Normal',np.nan,'Humid']
})
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
import seaborn as sns
sns.heatmap(df.isnull())
```

```python
ndf = df.dropna()
ndf
```

```python
sns.heatmap(ndf.isnull())
```

```python
fuldf = df.dropna(how="all")
```

```python
fuldf
```

```python
df
```

```python
ffill = fuldf.fillna(method="ffill")
```

```python
ffill
```

```python
fuldf
```

```python
meanv = fuldf["Humidity"].mean()
```

```python
meanv
```

```python
fuldf["Humidity"].fillna(meanv)
```

```python
fuldf
```

```python
fuldf["Humidity"].fillna(meanv, inplace=True)
```

```python
fuldf
```

```python
fuldf.drop_duplicates()
```
